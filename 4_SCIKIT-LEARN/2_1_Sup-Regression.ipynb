## Linear Regression - Detailed Notes

### **What is Linear Regression?**
- A supervised learning algorithm used to predict continuous numerical values
- Finds the best-fit straight line through data points
- Assumes a linear relationship between input (X) and output (y)

### **Mathematical Formula**
- **y = m × x + b**
  - `y` = predicted output value
  - `m` = slope (rate of change)
  - `x` = input feature
  - `b` = y-intercept (constant)

### **How It Works**
1. **Collect Training Data**: Historical data with known inputs and outputs
2. **Fit the Model**: Algorithm finds the best line by minimizing error
3. **Make Predictions**: Use the line equation to predict new values

### **When to Use Linear Regression**
- Predicting house prices based on area
- Estimating test scores based on study hours
- Forecasting sales based on advertising spend
- Any situation with linear relationships

### **Advantages**
- Simple and easy to understand
- Fast to train and predict
- Works well with small to medium datasets
- Provides interpretable results

### **Disadvantages**
- Assumes linear relationship (may not fit complex patterns)
- Sensitive to outliers
- Cannot capture non-linear patterns
- Poor performance if data doesn't follow linear trend

### **Key Metrics**
- **R² Score**: How well the line fits the data (0-1, higher is better)
- **Mean Squared Error (MSE)**: Average of squared differences between predicted and actual values
- **Mean Absolute Error (MAE)**: Average absolute differences between predictions and actual values

```python
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3], [4], [5]]
y = [40, 50, 60, 70, 80]
X_train = X[:3]
y_train = y[:3] 
model = LinearRegression()
model.fit(X_train, y_train)
hours = float(input("Enter number of hours studied: "))
predictions = model.predict([[hours]])
print(f"Predicted score for studying {hours} hours: {predictions[0]}")
```
