# Label Encoding

Definition: Label Encoding is a preprocessing technique that converts categorical (text/string) data into numerical values by assigning a unique integer to each category. It maps each unique category to a number, typically starting from 0. This is useful for converting categorical features into a format that machine learning algorithms can understand.

Use Cases:

- Binary classification: convert Yes/No, Male/Female, etc. to 0/1
- Multi-class categorical features: convert colors, cities, grades, etc. to 0, 1, 2, ...
- Ordinal features: when there's a natural ordering (Good > Better > Best)

Important Notes:

- Label Encoding assigns arbitrary numbers (0, 1, 2, ...) with no inherent relationship
- For nominal categories (no order), consider One-Hot Encoding to avoid implying order
- For ordinal categories (with natural order), Label Encoding preserves the relationship
- Scikit-learn's `LabelEncoder` automatically assigns integers based on alphabetical order of categories

````python
from sklearn.preprocessing  import LabelEncoder
import pandas as pd

df = pd.read_csv('sample_data.csv')  # Assuming a CSV file with categorical data
print("Original DataFrame:")
df_label = df.copy()
print(df_label.head())
le = LabelEncoder()

df_label['Gender_encoded'] = le.fit_transform(df_label['Gender'])
df_label['Passed_encoded'] = le.fit_transform(df_label['Passed'])

print("\nDataFrame after Label Encoding:")
print(df_label[['Name', 'Gender', 'Gender_encoded', 'Passed', 'Passed_encoded']].head())

````

# get_dummies (One-Hot Encoding)

Definition: `pd.get_dummies()` is a pandas function that performs One-Hot Encoding on categorical data. It converts categorical variables into dummy/indicator variables (binary columns), creating a new binary column for each unique category. Each row gets a 1 in the column corresponding to its category and 0s in all other category columns.

Use Cases:

- Converting nominal (non-ordinal) categorical features for machine learning models
- Preparing data for linear models (linear regression, logistic regression) that don't handle categories well
- Converting multiple categorical columns simultaneously
- Creating dummy variables for statistical analysis

Key Parameters:

- **`data`**: DataFrame or Series to process
- **`columns`**: column names to encode (if None, encodes all object/category columns)
- **`prefix`**: prefix to add to new column names (string or dict)
- **`prefix_sep`**: separator between prefix and category value (default '_')
- **`drop_first`**: if True, drops first category to avoid multicollinearity (useful for linear models)
- **`dtype`**: data type for dummy variables (default uint8, can use int for int type)

Important Notes:

- One-Hot Encoding creates more columns than Label Encoding (one per category)
- Use `drop_first=True` with linear models to avoid the "dummy variable trap" (perfect multicollinearity)
- Better for nominal features where no ordinal relationship exists
- Tree-based models (Random Forest, XGBoost) typically work well with Label Encoding

# StandardScaler (Feature Scaling)

Definition: StandardScaler is a preprocessing technique from scikit-learn that standardizes features by removing the mean and scaling to unit variance. It transforms each feature to have a mean of 0 and a standard deviation of 1, following the formula: `z = (X - mean) / std_dev`. This is also called Z-score normalization or standardization.

Formula:

```
z = (X - μ) / σ

where:
  X = original value
  μ = mean of the feature
  σ = standard deviation
  z = scaled value
```

Use Cases:

- Algorithms sensitive to feature magnitude: KNN, K-means, SVM, neural networks
- Linear models: linear regression, logistic regression (improves convergence)
- Distance-based algorithms that use Euclidean distance
- When features have different units or scales

Important Notes:

- StandardScaler centers data around 0 with spread of ±3σ (most values in [-3, 3] range)
- **Must fit on training data only**, then transform training and test data
- Preserves outliers (unlike Min-Max scaling which bounds to [0,1])
- Works well with features that follow a normal distribution
- Use after splitting into train/test sets to prevent data leakage

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
from sklearn.model_selection import train_test_split

data = { 'StudyHours': [2, 3, 5, 7, 9, 11, 4, 6, 8, 10],
         'TestScore': [50, 55, 65, 70, 75, 80, 60, 68, 72, 78] }
df = pd.DataFrame(data)

scaler = StandardScaler()
scaler = scaler.fit_transform(df)

print("Standardized Data:")
print(pd.DataFrame(scaler, columns=df.columns))
````

# MinMaxScaler (Normalization)

Definition: MinMaxScaler is a preprocessing technique from scikit-learn that rescales features to a fixed range, typically [0, 1]. It transforms each feature by subtracting the minimum value and dividing by the range (max - min), mapping all values linearly to the interval [0, 1]. This is also called min-max normalization or feature scaling.

Formula:

```
X_scaled = (X - X_min) / (X_max - X_min)

where:
  X = original value
  X_min = minimum value of the feature
  X_max = maximum value of the feature
  X_scaled = scaled value in [0, 1]
```

Use Cases:

- Neural networks and deep learning (expects inputs in [0, 1] range)
- Algorithms where you need bounded output: bounded feature importance
- When you need all values strictly between 0 and 1
- Image processing (pixel values typically normalized to [0, 1])
- When features have known minimum and maximum bounds

Important Notes:

- MinMaxScaler scales all values to [0, 1] range (or any specified range)
- Sensitive to outliers: outliers stretch the scale, compressing other values
- **Must fit on training data only**, then transform training and test data
- Best used when you know the actual min/max values of your data
- All values are bounded (min=0, max=1 after scaling)
- Use after splitting into train/test sets to prevent data leakage

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
from sklearn.model_selection import train_test_split

data = { 'StudyHours': [2, 3, 5, 7, 9, 11, 4, 6, 8, 10],
         'TestScore': [50, 55, 65, 70, 75, 80, 60, 68, 72, 78] }
df = pd.DataFrame(data)


min_max_scaler = MinMaxScaler()
min_max_scaled = min_max_scaler.fit_transform(df)

print("Min-Max Scaled Data:")
print(pd.DataFrame(min_max_scaled, columns=df.columns))
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Example 2: Understanding scaler attributes and transformation
print('--- Example 2: Scaler Attributes and Transformation ---\n')

data = {
    'Height': [160, 165, 170, 175, 180],
    'Weight': [60, 65, 70, 75, 80]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Fit scaler
scaler = StandardScaler()
scaler.fit(df)

print("\nScaler Attributes:")
print(f"Mean: {scaler.mean_}")
print(f"Std Dev: {scaler.scale_}")

# Manual calculation for first row
print("\nManual transformation for first row [160, 60]:")
height_scaled = (160 - scaler.mean_[0]) / scaler.scale_[0]
weight_scaled = (60 - scaler.mean_[1]) / scaler.scale_[1]
print(f"Height: (160 - {scaler.mean_[0]:.2f}) / {scaler.scale_[0]:.2f} = {height_scaled:.4f}")
print(f"Weight: (60 - {scaler.mean_[1]:.2f}) / {scaler.scale_[1]:.2f} = {weight_scaled:.4f}")

# Using transform method
scaled_data = scaler.transform(df)
print("\nUsing scaler.transform():")
print(f"First row: {scaled_data[0]}")

# Inverse transform (convert back to original scale)
print("\nInverse Transform (back to original scale):")
original = scaler.inverse_transform(scaled_data)
print(original)
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Example 3: Proper workflow - fit on training data only
print('--- Example 3: Train/Test Split Workflow (IMPORTANT!) ---\n')

data = {
    'Feature1': [10, 20, 30, 40, 50, 60, 70, 80],
    'Feature2': [100, 200, 300, 400, 500, 600, 700, 800],
    'Target': [0, 1, 0, 1, 0, 1, 0, 1]
}
df = pd.DataFrame(data)

# Step 1: Split data FIRST
X = df[['Feature1', 'Feature2']]
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Training set:")
print(X_train)
print("\nTest set:")
print(X_test)

# Step 2: Fit scaler on training data ONLY
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Step 3: Transform test data using TRAINING scaler parameters
X_test_scaled = scaler.transform(X_test)

print("\nScaled Training Data:")
print(pd.DataFrame(X_train_scaled, columns=X.columns))

print("\nScaled Test Data:")
print(pd.DataFrame(X_test_scaled, columns=X.columns))

print("\nWhy fit on train only?")
print("- Prevents data leakage (test data shouldn't influence training preprocessing)")
print("- Scaler learns data distribution from training set")
print("- Test set is scaled using training set's mean and std")
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Example 2: Understanding scaler attributes and custom feature_range
print('--- Example 2: Scaler Attributes and Custom Range ---\n')

data = {
    'Temperature': [0, 10, 20, 30, 40],
    'Humidity': [20, 40, 60, 80, 100]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Fit scaler
scaler = MinMaxScaler()
scaler.fit(df)

print("\nScaler Attributes:")
print(f"Data Min: {scaler.data_min_}")
print(f"Data Max: {scaler.data_max_}")
print(f"Feature Range: {scaler.feature_range}")

# Transform to [0, 1]
scaled_01 = scaler.transform(df)
print("\nScaled to [0, 1]:")
print(pd.DataFrame(scaled_01, columns=df.columns))

# Custom feature_range: [-1, 1]
scaler_custom = MinMaxScaler(feature_range=(-1, 1))
scaled_custom = scaler_custom.fit_transform(df)
print("\nScaled to [-1, 1] with custom feature_range:")
print(pd.DataFrame(scaled_custom, columns=df.columns))

# Inverse transform
print("\nInverse Transform (back to original):")
original = scaler.inverse_transform(scaled_01)
print(original)
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Example 3: Proper workflow - fit on training data only
print('--- Example 3: Train/Test Split Workflow (IMPORTANT!) ---\n')

data = {
    'Feature1': [10, 20, 30, 40, 50, 60, 70, 80],
    'Feature2': [5, 15, 25, 35, 45, 55, 65, 75],
    'Target': [0, 1, 0, 1, 0, 1, 0, 1]
}
df = pd.DataFrame(data)

# Step 1: Split data FIRST
X = df[['Feature1', 'Feature2']]
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print("Training set:")
print(X_train)
print(f"\nTest set:")
print(X_test)

# Step 2: Fit scaler on training data ONLY
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Step 3: Transform test data using TRAINING scaler parameters
X_test_scaled = scaler.transform(X_test)

print("\nScaled Training Data (fit on this data):")
print(pd.DataFrame(X_train_scaled, columns=X.columns))

print("\nScaled Test Data (transformed using training scaler):")
print(pd.DataFrame(X_test_scaled, columns=X.columns))

print("\nScaler Data Min/Max from TRAINING data:")
print(f"Min: {scaler.data_min_}")
print(f"Max: {scaler.data_max_}")

print("\nWhy fit on train only?")
print("- Prevents data leakage (test data shouldn't influence training preprocessing)")
print("- Scaler learns min/max from training set")
print("- Test set is scaled using training set's min/max")
print("- Test values may not be exactly in [0, 1] if they're outside training range")
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Example 4: Sensitivity to outliers
print('--- Example 4: MinMaxScaler Sensitivity to Outliers ---\n')

# Data WITHOUT outlier
data_normal = {
    'Values': [10, 20, 30, 40, 50]
}
df_normal = pd.DataFrame(data_normal)

# Data WITH outlier
data_outlier = {
    'Values': [10, 20, 30, 40, 50, 1000]  # 1000 is extreme outlier
}
df_outlier = pd.DataFrame(data_outlier)

# Scale normal data
scaler1 = MinMaxScaler()
scaled_normal = scaler1.fit_transform(df_normal)

print("Original normal data (no outlier):")
print(df_normal.values.flatten())
print("Scaled (min=10, max=50):")
print(scaled_normal.flatten())
print(f"Min: {scaled_normal.min():.4f}, Max: {scaled_normal.max():.4f}")

# Scale data with outlier
scaler2 = MinMaxScaler()
scaled_outlier = scaler2.fit_transform(df_outlier)

print("\n\nOriginal data with outlier:")
print(df_outlier.values.flatten())
print("Scaled (min=10, max=1000):")
print(scaled_outlier.flatten())
print(f"Min: {scaled_outlier.min():.4f}, Max: {scaled_outlier.max():.4f}")

print("\n\nObservation:")
print("- Without outlier: values spread across [0, 1]")
print("- With outlier: normal values compressed to very small values (0.0-0.045)")
print("- The range is stretched to accommodate the outlier")
print("- This can harm model performance if outliers are not handled first!")
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Example 5: Practical use case - Image pixel normalization
print('--- Example 5: Practical Use Case - Image Pixel Normalization ---\n')

# Simulating image pixel values (0-255 range)
print("Use Case: Normalizing image pixel values for neural networks\n")

pixel_data = {
    'Red': [0, 50, 100, 150, 200, 255],
    'Green': [0, 75, 150, 225, 255, 255],
    'Blue': [255, 200, 150, 100, 50, 0]
}
df_pixels = pd.DataFrame(pixel_data)
print("Original pixel values (0-255):")
print(df_pixels)

# Normalize to [0, 1] for neural networks
scaler = MinMaxScaler()
pixels_normalized = scaler.fit_transform(df_pixels)

df_norm = pd.DataFrame(pixels_normalized, columns=df_pixels.columns)
print("\nNormalized pixel values (0-1):")
print(df_norm)

print("\nWhy normalize pixels to [0-1]?")
print("- Neural networks work better with small input values")
print("- Improves training stability and convergence")
print("- Pixel range [0, 255] is well-suited for MinMaxScaler")
print("- Values are bounded and interpretable")

# Demonstrate with a simple example
print("\n\nOther practical examples:")
print("- Stock prices (scaled to comparable ranges)")
print("- Percentage values (already 0-100, scale to 0-1)")
print("- Age ranges (e.g., 0-120 scaled to 0-1)")
print("- Distance/feature measurements across different scales")
````
