Subplot Functions in Matplotlib
Subplot Functions are methods used to create multiple plots within a single figure. They allow you to arrange and display multiple axes (subplots) in a grid layout.

Key Subplot Functions:
plt.subplot(rows, cols, index) - Creates a subplot in a grid layout. Parameters specify rows, columns, and the position (index) of the current subplot.

Example: plt.subplot(2, 2, 1) creates a 2x2 grid and activates the 1st subplot.
plt.subplots(rows, cols) - Creates a figure with multiple subplots at once and returns both the figure and axes objects.

Example: fig, axes = plt.subplots(2, 2) creates a 2x2 grid of subplots.
fig.add_subplot(rows, cols, index) - Object-oriented method to add a subplot to a figure.

Example: ax = fig.add_subplot(2, 2, 1)
Common Use Cases:
Comparing multiple datasets side-by-side
Displaying different plot types together
Creating dashboard-style visualizations
Organizing related plots in a structured layout

```python
import matplotlib.pyplot as plt

# Example usage
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd subplot
plt.bar(x, y, color='orange')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Bar Chart')

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st subplot
plt.plot(x, y, color='green', marker='o')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Chart')

#plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

#fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))
ax1.plot(x, y, color='green', marker='o')
ax1.set_xlabel('X-axis Label')  
ax1.set_ylabel('Y-axis Label')
ax1.set_title('Line Chart')
ax2.bar(x, y, color='orange')
ax2.set_xlabel('X-axis Label')
ax2.set_ylabel('Y-axis Label')
ax2.set_title('Bar Chart')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

#fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(8, 6))
ax[0].plot(x, y, color='green', marker='^')
ax[0].set_xlabel('X-axis Label')  
ax[0].set_ylabel('Y-axis Label')
ax[0].set_title('Line Chart')
ax[1].bar(x, y, color='orange')
ax[1].set_xlabel('X-axis Label')
ax[1].set_ylabel('Y-axis Label')
ax[1].set_title('Bar Chart')
fig.suptitle('Line and Bar Charts Side by Side')
fig.tight_layout()
plt.tight_layout()
plt.show()
```
