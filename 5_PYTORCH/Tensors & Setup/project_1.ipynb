## Backward

```python
import torch

x = torch.tensor([2.0])  # Input - does not require gradients or be learnable
w = torch.tensor([3.0], requires_grad=True)  # Weight - requires gradients and is learnable
y_pred = w * x
y_true = torch.tensor([10.0])  # True value

loss = (y_pred - y_true) ** 2  # Mean squared error
print(f'Predicted value: {y_pred}')  # Should be 6 since x is 2 and w is 3
print(f'Loss: {loss}')  # Should be 16 since the difference is 4
print(f'Actual value: {y_true}')  # Should be 10
loss.backward()  # Compute gradients
print(f'Gradient of w: {w.grad}')  # Should be -8 since the loss is 16 and w is 3
print(f'Gradient of x: {x.grad}')  # Should be None since x does not require gradients
```

## Grad
- w.grad > 0 -> decrese value
- w.grad < 0 -> increase value

## Gradient Decent

- new_value = old_value - (learn_Rate * gradient)
- learn_grade = 0.1 * advice


```python
# input fixed
x = torch.tensor([2.0])  # Input - does not require gradients or be learnable

# learnable parameter
w = torch.tensor([3.0], requires_grad=True)  # Weight - requires gradients and is learnable

#correct answer
y_true = torch.tensor([10.0])  # True value

#learning rate
learning_rate = 0.1


y_pred = w * x
loss = (y_pred - y_true) ** 2  # Mean squared error

loss.backward()  # Compute gradients

print(f'Before update: w = {w.item()}, loss = {loss.item()}, gradient = {w.grad.item()}')
with torch.no_grad():  # Disable gradient tracking for the update step
    w -= learning_rate * w.grad  # Update w using the gradient

# IMPORTANT: After updating the parameter, we need to reset the gradients to zero before the next iteration, 
# otherwise the gradients will accumulate.
w.grad.zero_()  # Reset gradients to zero after the update
print(f'After update: w = {w.item()}, loss = {(w * x - y_true).item()}, gradient = {w.grad.item()}')

# After one update, w should be closer to 5, which would make the predicted value closer to 10 and reduce the loss.
# w- = 3 - (0.1 * (-16)) = 3 + (-1.6) = 4.6(After Update: 4.599999904632568)
```

## Training Loop
- Make Prediction
- Measure(loss)
- final
- fix

```python
x = torch.tensor([1.,2.,3.,4.])
y_true = torch.tensor([2.,4.,6.,8.]) # correct answer

# 
w = torch.tensor([0.0], requires_grad=True) # initial guess for w

learning_rate = 0.1 # learning rate - how much we want to update our parameters based on the gradients
epochs = 10 # number of iterations

for epoch in range(epochs):
    # step 1: forward pass - compute predicted y
    y_pred = w * x
    # step 2: compute loss
    loss = torch.mean((y_pred - y_true) ** 2)  # Mean squared error
    loss.backward()  # Compute gradients
    
    with torch.no_grad():  # Disable gradient tracking for the update step
        w -= learning_rate * w.grad  # Update w using the gradient
        # w.grad.zero_()  # Reset gradients to zero after the update
    
    print(f'Epoch {epoch+1}: w = {w.item():.4f}, loss = {loss.item():.4f}')
```
